```python
import pandas as pd
import os
```

```python
df = pd.read_csv('boe_codes.csv', index_col=0)
codes = list(df['BOE_codes'].values)
boe_folder = 'boe_data'
```

```python
len(set(codes))
```

```python
files = os.listdir(boe_folder)
combined = pd.DataFrame()
for f in files:
    path = os.path.join(boe_folder, f)
    df = pd.read_csv(path, index_col=0)
    df.index = pd.to_datetime(df.index, format="%d %b %Y").date
    for a in df.columns:
        if a not in combined.columns:
            combined = pd.concat([combined, df[a]], axis=1)
len(combined.columns)
```

```python
combined
```

```python
combined.to_csv('BOE_data.csv')
```

```python
missing = []
for a in codes:
    if a not in combined.columns:
        missing.append(a)
len(missing)
```
