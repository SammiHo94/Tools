```python
import os
import pandas as pd
```

```python
base_df = pd.read_csv('data/factset_cons_data_xignite.csv')
base_df['date'] = pd.to_datetime(base_df['date'], format="%Y-%m-%d").dt.date
base_df.head()
```

```python
base_df2 = pd.read_csv('data/factset_ratio_beta_base.csv')
base_df2['date'] = pd.to_datetime(base_df2['date'], format="%Y-%m-%d").dt.date
base_df2 = base_df2[base_df2['date']>base_df['date'].values[-1]]
base_df2.head()
```

```python
base_df = pd.concat([base_df, base_df2])
base_df = base_df.drop(['write_time'], axis=1)
base_df.head()
```

```python
base_df.tail()
```

```python
new_df = pd.read_csv('data/220228T033519/factset_cons_data.csv')
new_df['date'] = pd.to_datetime(new_df['date'], format="%Y-%m-%d").dt.date
new_df
```

```python
td = list(set(new_df.dropna()['date']))
td.sort()
```

```python
dd = {
    'xignite': base_df[base_df['date'].isin(td)],
    'investingcom': new_df[new_df['date'].isin(td)]
}
df = pd.concat(dd.values(), keys=dd.keys(),axis=1)
df
```

```python
df.to_csv('data/comparison.csv')
```

```python
nstocks = 20
result = pd.Series()
acc = pd.Series()

for dt in td:
    base_symbol = list(base_df[base_df['date'] == dt]['code'].values)
    base_symbol.sort()
    new_symbol = list(new_df[new_df['date'] == dt]['code'].values)
    new_symbol.sort()
#     print(base_symbol)
#     print(new_symbol)
    a_set = set(base_symbol)
    b_set = set(new_symbol)
    
    result[dt] = len(a_set.intersection(b_set))
    acc[dt] = len(a_set.intersection(b_set))/nstocks
result.name = 'common asset number'
acc.name = '% of total assets'
res = pd.concat([result, acc], axis=1)
res
```
