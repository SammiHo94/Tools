```python
import requests
import pandas as pd
```

```python
api_key = 'eaed9960de9b4fd688dd3a7523108179'
file_type = 'json'
```

### get releases category id

```python
req_category_url = 'https://api.stlouisfed.org/fred/releases?api_key={}&file_type={}'.format(api_key, file_type)
r = requests.get(req_category_url)
releases = r.json()['releases']
category_df = pd.DataFrame(releases)
category_df = category_df[['id', 'name']]
category_df = category_df.set_index('id')
category_df.head()
```

## get the series id string of each category

```python
req_series_url = 'https://api.stlouisfed.org/fred/category/series?category_id={}&api_key={}&file_type={}'
series_df = pd.DataFrame()
failed_id = []
for category_id in category_df.index:
#     print("**********************************************************")
    r = requests.get(req_series_url.format(category_id, api_key, file_type))
    print('category_id: ', category_id, ' r.status_code: ', r.status_code)
#     print(r.json().keys())
    if r.status_code == 200 and 'seriess' in r.json().keys():
        seriess = r.json()['seriess']
        series = pd.DataFrame(seriess)
        series['category_id'] = category_id
        series['category_name'] = category_df.loc[category_id, 'name']
#         print(series)
        series_df = pd.concat([series_df, series])
    else:
        failed_id.append(category_id)
series_df.head()
```

```python
series_df.to_csv('FRED_series.csv', index=False)
```
