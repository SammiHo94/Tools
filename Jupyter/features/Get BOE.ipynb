```python
import xlrd
import os
import requests
import wget
import numpy as np
import pandas as pd
from urllib.parse import urlparse, parse_qs
```

BOE_code -> unzip the file which downloaded from

https://www.bankofengland.co.uk/-/media/boe/files/statistics/bankstats-latest-tables.zip?la=en&hash=9E14AD11C713DDB25B3A2AC74D69847AB69924CD

```python
folder = '.\BOE_code'
files = os.listdir(folder)
# boe_codes = []
df = pd.DataFrame()
for f in files:
    fn = os.path.join(folder, f)
#     print("======================================", fn)
    wb = xlrd.open_workbook(fn)
    for sheet in wb.sheets():
        unit = sheet.cell(2, 0).value
#         print(unit)
        sub_code = []
        for link in sheet.hyperlink_list:
            url = link.url_or_path
            parsed_url = urlparse(url)
            captured_value = parse_qs(parsed_url.query)
            if 'EC' in captured_value.keys():
                sub_code += captured_value['EC']
        sub_df = pd.DataFrame({'BOE_codes': sub_code, 'unit': unit})
        df = pd.concat([df, sub_df])
```

```python
df = df.drop_duplicates(['BOE_codes'], keep='first')
df
```

```python
df.to_csv('boe_codes.csv')
```
