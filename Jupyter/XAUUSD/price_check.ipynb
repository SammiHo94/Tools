```python
import os
import pandas as pd
```

```python
data_folder = './data/every5mins/'
assets = ['XAUUSD', 'XAGUSD']
```

```python
gold = pd.DataFrame()
silver = pd.DataFrame()

price_dict = {}
for f in os.listdir(data_folder):
    # print("==========================================", f)
    asset = f.split('_')[0]
    d = f.split("T")[0].split('_')[-1]#.split("T")[0]
    t = f.split("T")[-1][:-4]
    # print(asset, t)

    ser = pd.read_csv(os.path.join(data_folder, f), index_col=0)['close']
    if d == '2022-08-05':
        ser.name = d
    else:
        ser.name = t

    if asset not in price_dict.keys():
        price_dict[asset] = pd.DataFrame()
    price_dict[asset] = pd.concat([price_dict[asset], ser], axis=1)
```

```python
tmp = pd.DataFrame()
for asset, df in price_dict.items():
    df.index = pd.to_datetime(df.index)
    tt = df.loc['2022-08-04', :]
    tt.T.name = asset
    tmp = pd.concat([tmp, tt], axis=1)
tmp
```
