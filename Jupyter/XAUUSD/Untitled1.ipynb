```python
# period = 'W'
# datelist = pd.Series(pd.to_datetime(df.index), index=pd.to_datetime(df.index))

def period_return(df):
    pct_df = pd.DataFrame()
    for i in range(1, len(df)):
        sub_pct = df.pct_change(i)
        sub_pct = sub_pct.dropna()
        pct_df = pd.concat([pct_df, sub_pct.iloc[[0], :]])
    return pct_df
    

def split_period(datelist, period):
    datelist = datelist.sort_index()
    period_index = datelist.index.to_period(period)
    period_start = datelist.groupby(period_index).nth(0)
    period_start.name = 'start'
    period_end = datelist.groupby(period_index).nth(-1)
    period_end.name = 'end'
    period_df = pd.concat([period_start, period_end], axis=1)
    return period_df
```
