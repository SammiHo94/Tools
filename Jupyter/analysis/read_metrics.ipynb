```python
import pandas as pd
import os

import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings("ignore")
```

```python
metrics = {}
```

```python
cases_folders = [
    './result/2022/cashON_signal/result/long_hist/corr_threshold_longhist_v1/',
    # './result/2022/cashON_signal/result/long_hist/corr_threshold_longhist_v2/',
    # './result/2022/cashON_signal/result/long_hist/corr_threshold_longhist_v3/',
]

periods = ['full', '(1987, 2007)', '(2008, 2022)']

for cases_folder in cases_folders:
    print("============================================================")
    case = cases_folder.rsplit("/", 2)[1]#[:-1]
    print(case)
    # break
    metrics[case] = {}
    save_folder = os.path.join(cases_folder, 'metrics')
    for p in periods:
        print(p)
        metrics_foldor = os.path.join(save_folder, "{}.csv".format(p))
        if os.path.exists(metrics_foldor):
            df = pd.read_csv(metrics_foldor, index_col=0)
            df = df.sort_values('sharpe_ratio', ascending=False)
            metrics[case][p] = df

            ll = ['no_3vote'] + list(df.index[:10])

            ll = ['no_3vote', 
                '(corr_avg_abs__lb15__out-of0.24&0.31)&(corr_avg_abs_pct__lb20__out-of-0.04&0.04)', # -0.007847
                '(corr_avg_abs__lb15__out-of0.26&0.31)&(corr_avg_abs_pct__lb20__<-0.04)', # -0.011749
                '(corr_avg_abs__lb15__out-of0.26&0.31)&(corr_avg_abs_pct__lb20__out-of-0.04&0.02)' # -0.002585
            ]
            display(df.loc[ll, :].sort_values('sharpe_ratio', ascending=False))
        else:
            print("No file {}".format(metrics_foldor))
```
